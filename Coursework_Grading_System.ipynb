# Project Title: Grading System for Coursework

Project Specification:
This grading system is a service for every professor in our school to simplify the marking process for students' coursework. Our system allow professors to input a raw file that contains all students' details and marks then proceed to calculate and generate a output file for further processes. Weightage or percentage of the assignment and test can also be input by professors to ensure different courses can be calculated too. This program not only saves valuable time but ensures consistent and fair grading practices.

# Problem Analysis

The problem every professors face are the calculation and grading of thousands students' coursework assessment. This process consumes a lot of time and may cause careless mistake. A program would help professor to ease this work. Professors just need to input a raw file contain all the students' data then a new file will be graded and output to professors.

# Process


---
#There are four main function in this program:

1. Read function
* Open file provided by professor then collect data and assign into the correct array

2. Write function
* Output to a another file that is finish grading with a similar format

3. Calculate function
* Calculate marks accurately with the weightage input by professor

4. Grade function
* Give grade to students with consistent system according to the total coursework mark

* An output file (CPT111_CWMarks_Output.txt) produced for professors' further process

2. Invalid input
* Any number start with 1 (Ex: 100) will continue the program.

# Programming Design & Solution

**PSEUDOCODE**

    Include necessary library
    FUNCTION read(inFile, student, matricNo, test1, test2, assign1, assign2)
      OPEN inFile using input mode
      READ the first line (header) from inputFile

      FOR i = 0 TO student - 1
          READ matricNo[i], test1[i], test2[i], assign1[i], assign2[i] FROM inFile
      END FOR

      CLOSE inFile

      PRINT "This is all the students' data collected in the file"
      FOR i = 0 TO student - 1
          PRINT "Matric Number: ", matricNo[i], " Test 1: ", test1[i], " Test 2: ", test2[i], " Assignment 1: ", assign1[i], " Assignment 2: ", assign2[i]
      END FOR
    END FUNCTION

    FUNCTION write(outFile, student, matricNo, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade)
      OPEN outFile using output mode
      WRITE "Matric Number,Test 1,Test1_Weightage,Test 2,Test2_Weightage,Assignment 1,Assignment1_Weightage,Assignment 2,Assignment2_Weightage,Total CW,over 100,Grade" TO outputFile

      FOR i = 0 TO student - 1
          WRITE matricNo[i], ",", test1[i], ",", weightTest1[i], ",", test2[i], ",", weightTest2[i], ",", assign1[i], ",", weightAssign1[i], ",", assign2[i], ",", weightAssign2[i], ",", totalCw[i], ",", cw100[i], ",", grade[i] TO outFile
      END FOR

      CLOSE outFile

      PRINT "Students' coursework have been graded and output to a new file (CPT111_CWMarks_Output.txt)!"
    END FUNCTION

    FUNCTION calculate(student, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade)
      DECLARE pTest1, pTest2, pAssign1, pAssign2, totalPercentage

      PRINT "Enter the percentage for Test 1, Test 2, Assignment 1 and Assignment 2 (Ex: 10 10 15 15):"
      INPUT pTest1, pTest2, pAssign1, pAssign2

      WHILE pTest1 <= 0 or pTest2 <= 0 or pAssign1 <= 0 or pAssign2 <= 0
        print "Invalid Input"

      END WHILE

      totalPercentage = pTest1 + pTest2 + pAssign1 + pAssign2

      FOR i = 0 TO student - 1
          weightTest1[i] = test1[i] * (pTest1 / 100)
          weightTest2[i] = test2[i] * (pTest2 / 100)
          weightAssign1[i] = assign1[i] * (pAssign1 / 100)
          weightAssign2[i] = assign2[i] * (pAssign2 / 100)

          totalCw[i] = weightTest1[i] + weightTest2[i] + weightAssign1[i] + weightAssign2[i]
          cw100[i] = totalCw[i] * 100 / totalPercentage

          CALL calculateGrade(student, cw100, grade)
      END FOR
    END FUNCTION

    FUNCTION calculateGrade(student, cw100, grade)
      FOR i = 0 TO student - 1
          IF cw100[i] >= 80 THEN
            grade[i] = 'A'
          ELSE IF cw100[i] >= 70 THEN
            grade[i] = 'B'
          ELSE IF cw100[i] >= 60 THEN
            grade[i] = 'C'
          ELSE IF cw100[i] >= 50 THEN
            grade[i] = 'D'
          ELSE
            grade[i] = 'F'
          END IF
      END FOR
    END FUNCTION

    MAIN Function
      DECLARE inputMark, outputMark
      DECLARE totalCwmark, student
      DECLARE method

      PRINT "Hello Professor! Welcome using this program!"

      WHILE method != 'q' AND method != 'Q'
        PRINT "Press 1 to Calculate coursework"
        PRINT "or enter q/Q to quit this program"
        READ method

        SWITCH method
            CASE '1':
                PRINT "Enter the total marks for all the coursework:"
                READ totalCwmark
                PRINT "Enter the number of students:"
                READ student

                DECLARE matricNo[student], test1[student], test2[student], assign1[student], assign2[student], grade[student]
                DECLARE weightTest1[student], weightTest2[student], weightAssign1[student], weightAssign2[student], totalCw[student], cw100[student]

                CALL read(inputMark, student, matricNo, test1, test2, assign1, assign2)
                CALL calculate(student, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade)
                CALL write(outputMark, student, matricNo, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade)

                BREAK
        END SWITCH

        IF method == 'q' OR method == 'Q'
            PRINT "Thanks for using this program professor!"
            BREAK
        END IF
    END WHILE

    RETURN 0
    END FUNCTION

**The implementation of the program is given in the following code.**

```python
%%writefile calculateMarks.cpp

#include <iostream>
#include <fstream>
#include <string>
#include <iomanip>
using namespace std;

//all function prototype
void read(const string inFile, int student, int matricNo[], int* test1, int* test2, int* assign1, int* assign2);
			//function to read and assign variable with the data in the file
void write(const string outFile, int student, int* matricNo, int* test1, int* test2, int* assign1, int* assign2, double* weightTest1, double* weightTest2, double* weightAssign1, double* weightAssign2, double* totalCw, double* cw100, char* grade);
			//function that output to a new file
void calculate(int student, int* test1, int* test2, int* assign1, int* assign2, double* weightTest1, double* weightTest2, double* weightAssign1, double* weightAssign2, double* totalCw, double* cw100, char* grade);
			//function calculate each student's coursework mark
void calculateGrade(int student, double* cw100, char* grade);
			//function that grade based on total coursework mark each students

int main (){

	const string inputMark = "CPT111_CWMarks.txt";   		    //constant string variable for the input file name
	const string outputMark = "CPT111_CWMarks_Output.txt";	//constant string variable for the output file name

	int totalCwmark, student;	  //declare for user input
	char method;				        //declare for user choice

	cout << "Hello Professor! Welcome using this program!\n";	//welcome speech

	while(method != 'q' && method != 'Q'){
		cout << "***********************************\n";
		cout << "Press 1 to Calculate coursework\n";		//short menu for user
		cout << "or enter q/Q to quit this program\n";
		cout << "***********************************\n";
		cin >> method;

		if (method != '1' && method != 'q' && method != 'Q'){		//clear buffer for invalid input
        cout << "Invalid input!\n";
        cin.clear();
        cin.ignore(100, '\n');
        continue;  // Continue the loop to ask for input again
    }

		switch(method){

			case '1':
				{
				cout << "Enter the total marks for all the coursework: ";
				cin >> totalCwmark;
				cout << "Enter the number of students: ";
				cin >> student;

				int matricNo[student], test1[student], test2[student], assign1[student], assign2[student];    //declare the array needed
				char grade[student];
				double weightTest1[student], weightTest2[student], weightAssign1[student], weightAssign2[student], totalCw[student], cw100[student];
        //each data array is set with the size of number of student and suitable type of variable

        //call function to process
				read(inputMark, student, matricNo, test1, test2, assign1, assign2);
				calculate(student, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade);
				write(outputMark, student, matricNo, test1, test2, assign1, assign2, weightTest1, weightTest2, weightAssign1, weightAssign2, totalCw, cw100, grade);

				}
				break;
		}

		if(method == 'q' || method == 'Q'){
			cout << "Thanks for using this program professor!";
			break;
		}

	}
	return 0;
}

//each function definition
void read(const string inFile, int student, int matricNo[], int* test1, int* test2, int* assign1, int* assign2){
	ifstream inputFile;		//declare for open file using input mode
	string input;

	inputFile.open(inFile);
	if(inputFile){
		getline(inputFile, input);		//get fisrt line that is not data of coursework mark

		//start from second line loop to get students' data
		for(int i = 0; i < student; i++){
			inputFile >> matricNo[i];	//matric number
			inputFile.ignore();			//ignore the character ','
			inputFile >> test1[i];		//mark for test 1
			inputFile.ignore();
			inputFile >> test2[i];		//mark for test 2
			inputFile.ignore();
			inputFile >> assign1[i];	//mark for assignment 1
			inputFile.ignore();
			inputFile >> assign2[i];	//mark for assignment 2
			inputFile.ignore();
		}
	}

	inputFile.close();

	//display data to user
	cout << "This is all the students' data collected in the file\n";
	for(int i = 0; i < student; i++){
		cout << "Matric Number: " << matricNo[i] << " Test 1: " << test1[i] << " Test 2: " << test2[i]
		<< " Assignment 1: " << assign1[i] << " Assignment 2: " << assign2[i] << endl;
	}
}

void write(const string outFile, int student, int* matricNo, int* test1, int* test2, int* assign1, int* assign2, double* weightTest1, double* weightTest2, double* weightAssign1, double* weightAssign2, double* totalCw, double* cw100, char* grade){
	ofstream outputFile;	//declare for open file using output mode

	outputFile.open(outFile);
	outputFile << "Matric Number,Test 1,Test1_Weightage,Test 2,Test2_Weightage,Assignment 1,Assignment1_Weightage,Assignment 2,Assignment2_Weightage,Total CW,over 100,Grade\n";
	//output first line of the file that is content of each data

  //loop to output data calculated with the similar format as the input data
	for(int i = 0; i < student; i++){
		outputFile << fixed << showpoint << setprecision(2);
		outputFile << matricNo[i] << "," << test1[i] << "," << weightTest1[i] << ","<< test2[i] << "," << weightTest2[i] << ","
		<< assign1[i] << "," << weightAssign1[i] << "," << assign2[i] << "," << weightAssign2[i] << "," << totalCw[i] << ","
		<<  cw100[i] << "," << grade[i] <<endl;
	}
	outputFile.close();

	cout << "Students' coursework have been graded and output to a new file (CPT111_CWMarks_Output.txt)!\n";
	//display output file message
}

void calculate(int student, int* test1, int* test2, int* assign1, int* assign2, double* weightTest1, double* weightTest2, double* weightAssign1, double* weightAssign2, double* totalCw, double* cw100, char* grade){

	int pTest1, pTest2, pAssign1, pAssign2, totalPercentage;

	cout << "\nEnter the percentage for Test 1, Test 2, Assignment 1 and Assignment 2(Ex: 10 10 15 15): ";
	cin >> pTest1 >> pTest2 >> pAssign1 >> pAssign2;
	//user input for weightage of each test and assignment

	while(pTest1 <= 0 || pTest2 <= 0 || pAssign1 <= 0 || pAssign2 <= 0){
		cin.clear();
		cin.ignore(10, '\n');
		cout << "Invalid input! Please enter positive value: ";
		cin >> pTest1 >> pTest2 >> pAssign1 >> pAssign2;
	}

	totalPercentage = pTest1 + pTest2 + pAssign1 + pAssign2;

	//loop to calculate each students' coursework mark
	for(int i = 0; i < student; i++){
		weightTest1[i] = test1[i] * ((double)pTest1 / 100);
		weightTest2[i] = test2[i] * ((double)pTest2 / 100);
		weightAssign1[i] = assign1[i] * ((double)pAssign1 / 100);
		weightAssign2[i] = assign2[i] * ((double)pAssign2 / 100);
		//(double)make sure each coursework mark will multiply by decimal value instead of 0

		totalCw[i] = weightTest1[i] + weightTest2[i] + weightAssign1[i] + weightAssign2[i];
		cw100[i] = totalCw[i] * 100 / totalPercentage;

		calculateGrade(student, cw100, grade);
		//grade 'A' to 'F' for students p
	}
}

void calculateGrade(int student, double* cw100, char* grade){
	for(int i = 0; i < student; i++){	//loop to grade each students' coursework accordingly
		if(cw100[i] >= 80){
			grade[i] = 'A';
		}
		else if(cw100[i] >= 70){
			grade[i] = 'B';
		}
		else if(cw100[i] >= 60){
			grade[i] = 'C';
		}
		else if(cw100[i] >= 50){
			grade[i] = 'D';
		}
		else{
			grade[i] = 'F';
		}
	}
}
```
